```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
#Show graphs as cell output
%matplotlib inline
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False) 
```

### 1. Boxplot

```python
# Boxplot for revs per mile
sns.boxplot(data=car_data, x="Rev.per.mile", y='Manufacturer', order=["Audi", "Hyundai", "Suzuki", "Toyota"])
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?

Toyota

### 2. Histogram

```python
# Histogram of MPG
#Creating plotting data
features_to_plot = pd.DataFrame({'MPG.highway': car_data['MPG.highway'],
                                 'MPG.city': car_data['MPG.city']})
#plot it
features_to_plot.plot(kind='hist', ylabel="Average MPG")
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
The city

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure()
sns.lineplot(x='Wheelbase', y='Turn.circle', data=car_data)
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
This is a direct relationship or directly proportional relationship as the turing circle gets larger so does the the wheelbase.

### 4. Barplot

```python
# Barplot for horsepower
# Group data by'Type'
group_by_type = car_data.groupby(by=['Type'])

# Get the mean (average) for each type across all columns
car_data_mean = round(group_by_type.mean(numeric_only=True), 0) 

#Get a dataframe of what we need
df = pd.DataFrame({'Horsepower': car_data_mean['Horsepower']})

#Plot
df.plot(kind='bar', ylabel="Horsepower")
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
Yes it does.
